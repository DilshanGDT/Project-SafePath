```python
# Importing Libraries

import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Import some dashboard camera images

img1 = cv2.imread(r'images\road1.webp', 1)
img2 = cv2.imread(r'images\road4.jpg', 1)
img3 = cv2.imread(r'images\\road2.webp', 1)
```

```python
# Display the Original images
plt.figure(figsize=(15,15), num='test.img')

plt.subplot(131)
plt.title('Image 1', fontsize = 8)
plt.xticks([]), plt.yticks([])
plt.imshow(img1, 'gray')

plt.subplot(132)
plt.title('Image 2', fontsize = 8)
plt.xticks([]), plt.yticks([])
plt.imshow(img2, 'gray')

plt.subplot(133)
plt.title('Image 3', fontsize = 8)
plt.xticks([]), plt.yticks([])
plt.imshow(img3, 'gray')
```

```python
# Step 01 - Convert sample images to Gray Scale

# As everyone knows, the majority of the light colors used for road lanes are white and yellow. 
# In light of this, we can conclude that a grayscale image works better for identifying the brighter lines on a darker road. 
# Moreover, a grayscale image aids in the Canny edge detection at the edge detection stage. 

grayImg1 = cv2.cvtColor(img1, cv2.COLOR_RGB2GRAY)
grayImg2 = cv2.cvtColor(img2, cv2.COLOR_RGB2GRAY)
grayImg3 = cv2.cvtColor(img3, cv2.COLOR_RGB2GRAY)
```

```python
# Display the Converted images
plt.figure(figsize=(15,15), num='test.img')

plt.subplot(131)
plt.title('Gray Scale Image 1', fontsize = 8)
plt.xticks([]), plt.yticks([])
plt.imshow(grayImg1, 'gray')

plt.subplot(132)
plt.title('Gray Scale Image 2', fontsize = 8)
plt.xticks([]), plt.yticks([])
plt.imshow(grayImg2, 'gray')

plt.subplot(133)
plt.title('Gray Scale Image 3', fontsize = 8)
plt.xticks([]), plt.yticks([])
plt.imshow(grayImg3, 'gray')
```

```python
# Step 02 - Removing Noise of the Gray scale images (Image Pre-processing)

# Iutilize derivative-based methodologies in myproject. 
# Because derivatives are highly susceptible to noise, we should control the noise before doing any mathematical operations. 
# Using a small amounts of Gaussian blur to blur the image is one method of handling image noise. 
# The Gaussian kernel was applied to a 5x5 window size. 
# We can have smoother images with this little window since, the smaller the kernel size, the more undetectable the blur.

# Apply Gaussian blur to reduce noise
blurGrayImg15 = cv2.GaussianBlur(grayImg1, (5, 5), 0)
blurGrayImg13 = cv2.GaussianBlur(grayImg1, (3, 3), 0)

blurGrayImg25 = cv2.GaussianBlur(grayImg2, (5, 5), 0)
blurGrayImg23 = cv2.GaussianBlur(grayImg2, (3, 3), 0)

blurGrayImg35 = cv2.GaussianBlur(grayImg3, (5, 5), 0)
blurGrayImg33 = cv2.GaussianBlur(grayImg3, (3, 3), 0)
```

```python
# Display the Blurred images
plt.figure(figsize=(10,10), num='test.img')

plt.subplot(321)
plt.title('Blurred Gray Scale Image 1 (5x5 Gaussian Blur)', fontsize = 8)
plt.xticks([]), plt.yticks([])
plt.imshow(blurGrayImg15, 'gray')

plt.subplot(322)
plt.title('Blurred Gray Scale Image 1 (3x3 Gaussian Blur)', fontsize = 8)
plt.xticks([]), plt.yticks([])
plt.imshow(blurGrayImg13, 'gray')

plt.subplot(323)
plt.title('Blurred Gray Scale Image 2 (5x5 Gaussian Blur)', fontsize = 8)
plt.xticks([]), plt.yticks([])
plt.imshow(blurGrayImg25, 'gray')

plt.subplot(324)
plt.title('Blurred Gray Scale Image 2 (3x3 Gaussian Blur)', fontsize = 8)
plt.xticks([]), plt.yticks([])
plt.imshow(blurGrayImg23, 'gray')

plt.subplot(325)
plt.title('Blurred Gray Scale Image 3 (5x5 Gaussian Blur)', fontsize = 8)
plt.xticks([]), plt.yticks([])
plt.imshow(blurGrayImg35, 'gray')

plt.subplot(326)
plt.title('Blurred Gray Scale Image 3 (3x3 Gaussian Blur)', fontsize = 8)
plt.xticks([]), plt.yticks([])
plt.imshow(blurGrayImg33, 'gray')
```

```python
# Step 03 - Canny Edge Detection

lowThresholdValue = 50 
highThresholdValue = 150

cannyEdges15 = cv2.Canny(blurGrayImg15, lowThresholdValue, highThresholdValue)
cannyEdges13 = cv2.Canny(blurGrayImg13, lowThresholdValue, highThresholdValue)

cannyEdges25 = cv2.Canny(blurGrayImg25, lowThresholdValue, highThresholdValue)
cannyEdges23 = cv2.Canny(blurGrayImg23, lowThresholdValue, highThresholdValue)

cannyEdges35 = cv2.Canny(blurGrayImg35, lowThresholdValue, highThresholdValue)
cannyEdges33 = cv2.Canny(blurGrayImg33, lowThresholdValue, highThresholdValue)
```

```python
# Display the Blurred image
plt.figure(figsize=(10,10), num='test.img')

plt.subplot(321)
plt.title('Canny Edge Detected Image 1 (5x5)', fontsize = 8)
plt.xticks([]), plt.yticks([])
plt.imshow(cannyEdges15, 'gray')

plt.subplot(322)
plt.title('Canny Edge Detected Image 1 (3x3)', fontsize = 8)
plt.xticks([]), plt.yticks([])
plt.imshow(cannyEdges13, 'gray')

plt.subplot(323)
plt.title('Canny Edge Detected Image 2 (5x5)', fontsize = 8)
plt.xticks([]), plt.yticks([])
plt.imshow(cannyEdges25, 'gray')

plt.subplot(324)
plt.title('Canny Edge Detected Image 2 (3x3)', fontsize = 8)
plt.xticks([]), plt.yticks([])
plt.imshow(cannyEdges23, 'gray')

plt.subplot(325)
plt.title('Canny Edge Detected Image 3 (5x5)', fontsize = 8)
plt.xticks([]), plt.yticks([])
plt.imshow(cannyEdges35, 'gray')

plt.subplot(326)
plt.title('Canny Edge Detected Image 3 (3x3)', fontsize = 8)
plt.xticks([]), plt.yticks([])
plt.imshow(cannyEdges33, 'gray')
```

```python
# Step 04 - Masking the region of interest

def ROI(img):                                   # Make a function to masking
    height, width = img.shape[:2]
    mask = np.zeros_like(img)
    # Define a larger triangle region of interest with an even wider base
    triangle = np.array([[
        (0, height),                            # Bottom-left corner at the left edge
        (width, height),                        # Bottom-right corner at the right edge
        (int(0.5 * width), int(0.35 * height))  # Top corner remains the same
    ]], np.int32)
    cv2.fillPoly(mask, triangle, 255)
    masked_img = cv2.bitwise_and(img, mask)
    return masked_img

# Apply the region of interest mask to the edge-detected images
roiEdges15 = ROI(cannyEdges15)
roiEdges13 = ROI(cannyEdges13)
roiEdges25 = ROI(cannyEdges25)
roiEdges23 = ROI(cannyEdges23)
roiEdges35 = ROI(cannyEdges35)
roiEdges33 = ROI(cannyEdges33)
```

```python
# Display the Blurred image
plt.figure(figsize=(10,10), num='test.img')

plt.subplot(321)
plt.title('Canny Edge Detected Image 1 (5x5)', fontsize = 8)
plt.xticks([]), plt.yticks([])
plt.imshow(roiEdges15, 'gray')

plt.subplot(322)
plt.title('Canny Edge Detected Image 1 (3x3)', fontsize = 8)
plt.xticks([]), plt.yticks([])
plt.imshow(roiEdges13, 'gray')

plt.subplot(323)
plt.title('Canny Edge Detected Image 2 (5x5)', fontsize = 8)
plt.xticks([]), plt.yticks([])
plt.imshow(roiEdges25, 'gray')

plt.subplot(324)
plt.title('Canny Edge Detected Image 2 (3x3)', fontsize = 8)
plt.xticks([]), plt.yticks([])
plt.imshow(roiEdges23, 'gray')

plt.subplot(325)
plt.title('Canny Edge Detected Image 3 (5x5)', fontsize = 8)
plt.xticks([]), plt.yticks([])
plt.imshow(roiEdges35, 'gray')

plt.subplot(326)
plt.title('Canny Edge Detected Image 3 (3x3)', fontsize = 8)
plt.xticks([]), plt.yticks([])
plt.imshow(roiEdges33, 'gray')
```

```python
# Step 05 - Apply Hough Probabilistic Transform to detect lines

def houghTransform(image, original_image):
    lines = cv2.HoughLinesP(image, 1, np.pi / 180, threshold = 50, minLineLength = 100, maxLineGap = 10)
    if lines is not None:
        for line in lines:
            x1, y1, x2, y2 = line[0]
            cv2.line(original_image, (x1, y1), (x2, y2), (0, 255, 0), 2)
    return original_image

# Apply Hough Transform on the edge-detected images
houghImg1 = houghTransform(roiEdges15, img1.copy())
houghImg2 = houghTransform(roiEdges25, img2.copy())
houghImg3 = houghTransform(roiEdges35, img3.copy())

# We can avoid 3x3 gaussian blured images (image 4, 5 and 6) since 5x5 blurring looks fine.
houghImg4 = houghTransform(roiEdges13, img1.copy())
houghImg5 = houghTransform(roiEdges23, img2.copy())
houghImg6 = houghTransform(roiEdges35, img3.copy())

# Display the results (only 5x5 blurred images)
plt.figure(figsize=(15,15))
plt.subplot(131)
plt.title('Hough Transform Image 1', fontsize = 8)
plt.xticks([]), plt.yticks([])
plt.imshow(cv2.cvtColor(houghImg1, cv2.COLOR_BGR2RGB))

plt.subplot(132)
plt.title('Hough Transform Image 2', fontsize = 8)
plt.xticks([]), plt.yticks([])
plt.imshow(cv2.cvtColor(houghImg2, cv2.COLOR_BGR2RGB))

plt.subplot(133)
plt.title('Hough Transform Image 3', fontsize = 8)
plt.xticks([]), plt.yticks([])
plt.imshow(cv2.cvtColor(houghImg3, cv2.COLOR_BGR2RGB))

plt.show()
```

```python
# Step 06 - Draw the lines in the original images

def drawLinesOnOriginal(image, lines_image):
    final_image = cv2.addWeighted(image, 0.4, lines_image, 1, 0)
    return final_image

# Draw the detected lines on the original images
finalImg1 = drawLinesOnOriginal(img1, houghImg1)
finalImg2 = drawLinesOnOriginal(img2, houghImg2)
finalImg3 = drawLinesOnOriginal(img3, houghImg3)
```

```python
# Display the final results
plt.figure(figsize=(15, 15))

plt.subplot(131)
plt.title('Final Image 1 with Detected Lines', fontsize=8)
plt.xticks([]), plt.yticks([])
plt.imshow(cv2.cvtColor(finalImg1, cv2.COLOR_BGR2RGB))

plt.subplot(132)
plt.title('Final Image 2 with Detected Lines', fontsize=8)
plt.xticks([]), plt.yticks([])
plt.imshow(cv2.cvtColor(finalImg2, cv2.COLOR_BGR2RGB))

plt.subplot(133)
plt.title('Final Image 3 with Detected Lines', fontsize=8)
plt.xticks([]), plt.yticks([])
plt.imshow(cv2.cvtColor(finalImg3, cv2.COLOR_BGR2RGB))

plt.show()
```

```python
# Step 07 - Measuring Road Width
# Improved (Does not affect the mddlle line i.e. left & right lane divider)

def measureRoadWidth(image, metersPerPixel):
    # Convert the image to HSV color space to detect green lines
    hsvImage = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    
    # Define the range for the color green in HSV
    lowerGreen = np.array([40, 40, 40])
    upperGreen = np.array([70, 255, 255])
    
    # Create a mask for green color
    greenMask = cv2.inRange(hsvImage, lowerGreen, upperGreen)
    
    # Find contours in the mask
    contours, _ = cv2.findContours(greenMask, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    
    leftX = []
    rightX = []
    height = image.shape[0]
    width = image.shape[1]
    
    for contour in contours:
        for point in contour:
            x, y = point[0]
            if y > height * 0.8:  # Consider points near the bottom of the image
                # Classify points as left or right side of the road
                if x < width // 3:  # Leftmost third of the image
                    leftX.append(x)
                elif x > 2 * width // 3:  # Rightmost third of the image
                    rightX.append(x)

    if leftX and rightX:
        # Measure road width using the outermost points on the left and right side
        leftmostX = np.min(leftX)
        rightmostX = np.max(rightX)
        
        # Calculate the road width in pixels and convert it to meters
        roadWidthPixels = rightmostX - leftmostX
        roadWidthMeters = roadWidthPixels * metersPerPixel
    else:
        roadWidthMeters = 0
    
    return roadWidthMeters

# Example usage with a sample image and metersPerPixel value
image = finalImg2  # The image with road lines
metersPerPixel = 0.012  # Replace with the actual meters per pixel value for your images

# meters per pixels = actual image size in meters / image pixel value
roadWidth = measureRoadWidth(image, metersPerPixel)
print(f'Road Width: {roadWidth:.2f} meters')

# Display the image with detected green lines for verification
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title(f'Road Width: {roadWidth:.2f} meters')
plt.show()
```
